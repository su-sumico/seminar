# Folium基礎プロット

```python
import pandas as pd
import geopandas as gpd
import folium
```

```python
# 東京スカイツリーの緯度経度
lat = 35.71
lon = 139.81
```

```python
map = folium.Map(location=[lat, lon], zoom_start=16)
map
```

```python
#マーカー
folium.Marker(
    location=[lat, lon],
    icon=folium.Icon(color='red', icon='camera'),
    popup='東京スカイツリー',
    tooltip='Here!'
).add_to(map)
map
```

アイコンの種類はこちらを参照
https://glyphsearch.com/?library=glyphicons

```python
#circle
folium.Circle(
    location=[lat, lon],
    radius=100,
    color='blue',
    fill=True,
    fill_opacity=0.5
).add_to(map)
map
```

```python
#背景の地図
map = folium.Map(location=[lat, lon], zoom_start=16, tiles='cartodbdark_matter')
map
```

```python
map = folium.Map(location=[lat, lon], zoom_start=16, tiles='stamenterrain')
map
```

# オープンデータのPointプロット

```python
df = pd.read_csv('data/kokyoshisetsu.csv', encoding='shift-jis')
df
```

```python
df = df.loc[:, "施設名":"経度"]
geometry = gpd.points_from_xy(df['経度'], df['緯度'])
gdf = gpd.GeoDataFrame(df, geometry=geometry)
gdf
```

```python
gdf.crs=6668
```

```python
lat = 35.6291
lon = 139.7389
map=folium.Map(location=[lat, lon], zoom_start=13)
```

```python
folium.Choropleth(
    gdf
).add_to(map)
map
```

# ヒートマップ

```python
from folium.plugins import HeatMap
```

```python
lat = 35.6291
lon = 139.7389
map=folium.Map(location=[lat, lon], zoom_start=13)
```

```python
pointlist=[]
for index, row in df.iterrows():
    pointlist.append((row.緯度, row.経度))
```

```python
HeatMap(pointlist, radius=20, blur=5).add_to(map)
```
